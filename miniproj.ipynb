```python
from ultralytics import YOLO
model = YOLO("yolov8m.pt")
results=model.track(source="v.mp4",show=False,save = True,name="result",tracker="bytetrack.yaml",stream=True)
```

```python
from collections import defaultdict
class_counts= defaultdict(int)
tracked_ids = defaultdict(set)
for r in results:
    boxes = r.boxes  # Bounding boxes
    for box in boxes:
        conf = box.conf.item()  # Confidence score
        if conf > 0.65:  # Only consider detections with confidence > 65%
            cls = int(box.cls.item())  # Class of the detected object
            class_name = model.names[cls]  # Get the class name

            # Check if the box has a tracker_id
            if hasattr(box, 'id'):
                tracker_id = int(box.id.item())

                # If this ID hasn't been counted for this class, count it
                if tracker_id not in tracked_ids[class_name]:
                    class_counts[class_name] += 1
                    tracked_ids[class_name].add(tracker_id)

# Print the results
print("Unique object counts (confidence > 65%):")
for class_name, count in class_counts.items():
    if class_name in ['car','truck','bus','bicycle','motorcycle',]:
      print(f"{class_name}: {count}")

# Calculate and print total unique objects
total_unique_objects = sum(class_counts.values())
#print(f"Total unique objects: {total_unique_objects}")
```
